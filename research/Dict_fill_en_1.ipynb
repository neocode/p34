```python
import re

with open('EN_polarity_lexicon.txt') as f:
    dic_list = f.readlines()
dic_list = [x.strip('\n') for x in dic_list]
#print(dic_list)
dic = {}
for x in dic_list:   
    x = x.replace(',', '')    
    str_list = x.split()    
    #print(str_list)
    if len(str_list) == 3:
        if (not '_' in str_list[0]) and (float(str_list[2]) != 0):            
            dic[str_list[0]] = [str_list[1].upper(), float(str_list[2])]
print(dic)

with open ('dic_new_en.txt', 'w') as fp:
    for word in dic.keys():
        fp.write("%s %s %s\n" % (word, dic[word][0], dic[word][1]))        
#str_list =  [x.replace(',', '') for x in dic_list]
#print(str_list)
```
